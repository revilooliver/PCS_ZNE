```python
import os
import numpy as np
from qiskit import qpy
from utils.pce_vs_zne_utils import random_cliff_circs 

# Parameters
num_qubits_list = [12, 10, 8]
depth_list      = [20, 40, 60, 80, 100]
num_circs       = 20
tol             = 1e-8

# Folder to save
folder = "data_PCE_vs_ZNE/saved_circs"
os.makedirs(folder, exist_ok=True)

for n in num_qubits_list:
    pauli_string = "Z" * n
    for depth in depth_list:
        # Filename that encodes (n,depth,num_circs)
        fname = os.path.join(
            folder,
            f"rand_cliffs_n{n}_d{depth}_nc{num_circs}.qpy"
        )

        # If you want to skip regeneration when the file already exists:
        if os.path.exists(fname):
            print(f"→ Skipping n={n}, d={depth}: file exists.")
            continue

        # Generate the circuits
        circs, avg_cx = random_cliff_circs(
            num_qubits   = n,
            depth        = depth,
            num_circs    = num_circs,
            pauli_string = pauli_string,
            tol          = tol
        )

        # Save them
        with open(fname, "wb") as f:
            qpy.dump(circs, f)

        print(f"✔  Generated & saved {len(circs)} circuits for n={n}, d={depth} (avg CX={avg_cx:.1f}) → {fname}")
```

```python
circs[-1].draw("mpl", fold=-1)
```

```python
if os.path.exists(fname):
    with open(fname, "rb") as f:
        circs = qpy.load(f)
    # recompute CX‐counts (cheap)
    cx_counts = [circ.count_ops().get("cx", 0) for circ in circs]
    avg_cx = float(np.mean(cx_counts))
    print(f"Loaded {len(circs)} circuits from {fname} (avg CX = {avg_cx:.1f})")
```
